```python
#review
def word_freq(s):
    txt = s.str.lower().str.replace(r'\W+', ' ').str.cat(sep=' ')  # \W+ replace non-character and non-digital values
                                                                # cat concatenate strings with ' ' no line any more 
                                                                # becomes a big list 
    words = nltk.tokenize.word_tokenize(txt)  
    words_dist = nltk.FreqDist(w for w in words if w not in stopwords_en\
                                       and w not in string.punctuation and not w.isdigit()) 
    return words_dist
```

```python
words_dist_review=word_freq(review)  # review 
words_dist_title=word_freq(title)    # title
```

```python
from wordcloud import WordCloud
f, ax = plt.subplots(1, 2, figsize=(22, 10))
#plt.figure( figsize=(20,10) )
wordcloud = WordCloud(background_color='black',width=2000, height=1000).generate_from_frequencies(words_dist_title)#' '.join(rslt['Word']))
ax[0].imshow(wordcloud)
ax[0].axis("off")
#plt.show()

wordcloud = WordCloud(background_color='black',width=2000, height=1000).generate_from_frequencies(words_dist_review)#' '.join(rslt['Word']))
ax[1].imshow(wordcloud)
ax[1].axis("off")
#plt.show()
```

```python
def select_string(s,sel):
    if pd.isnull(s):
        return False
    else:
        if any(sel_w in s.lower().split() for sel_w in sel):
            return True
        else:
            return False
```
